```python
from bs4 import BeautifulSoup as bs
import requests as rq
import pandas as pd
site = rq.get('https://www.w3schools.com/xml/cd_catalog.xml')
soup = bs(site.text)
dat = soup.find_all('cd')
df = pd.DataFrame([i.text.split('\n')[1:-1] for i in dat], columns=['Наименование','Исполнитель','Страна','Компания','Стоимость','Год'])
print('Стоимость всех дисков = ', round(sum([float(i) for i in df['Стоимость']]), 2))
print('Последний год издания - ', max([int(i) for i in df['Год']]))

new_df =  df.replace(to_replace ="Polydor", value ="Poly") # Замена Polydor на Poly

# запись в XML-файл
xml_data = ['']
for column in new_df.columns:
    xml_data.append('<{}>'.format(column))  
    for field in df.index:
        
        xml_data.append('<{0}>{1}'.format(field, new_df[column][field]))
    xml_data.append(''.format(column))  
xml_data.append('')

with open('CompactDisk.xml', 'w+') as f:  
    for line in xml_data:
        f.write(line)
```

```python
from lxml import etree as et
import pandas as pd

root = et.Element('root')  # Create root element

for column in df.columns:
    entry = et.SubElement(root, column)  # Writing element
    for row in df.index:
        schild = row
        child = et.SubElement(entry, schild)  # Writing sub-elements
        child.text = str(df[column][schild])

xml_data = et.tostring(root)  # binary string
with open('coordinates.xml', 'w') as f:  # Write in XML file as utf-8
    f.write(xml_data.decode('utf-8'))
```
